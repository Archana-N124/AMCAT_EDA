## Research Question

#### 1. Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.

```python
cs_grads = df[df['Specialization'].str.contains('computer science', na=False)]

job_title = ['programmer analyst', 'software engineer', 'hardware engineer', 'associate engineer']

filt_df = cs_grads[cs_grads['Designation'].isin(job_title)]

salary_range = filt_df[(filt_df['Salary'] >= 250000) & (filt_df['Salary'] <= 300000)]

percentage = len(salary_range) / len(filt_df) * 100

print(f"Percentage of salaries in the range 2.5 - 3 lakhs: {percentage:.2f}%")
print(filt_df['Salary'].describe())

plt.figure(figsize=(8, 5))
sns.barplot(x='Designation', y='Salary', data=filt_df)
plt.title('Salary Distribution for Selected Job Titles')
plt.ylabel('Salary (in Lakhs)')
plt.xlabel('Job Title')
plt.show()
```

## Observation

####  The mean salary is approximately ₹331,339, which is above the 2.5-3 lakhs range mentioned in the claim.
####  The median salary (50% percentile) is ₹320,000, which also slightly higher than the upper limit of the mentioned claim.
####  Out of the filtered group, approximately 17.6% of individuals come within the 2.5-3 lakhs salary range.

#### From this we can conclude that, the claim holds true as 17.6% of individuals, the mean and median salaries of compueter science graduates for the specified job roles are higher than the mentined range in the claim.
