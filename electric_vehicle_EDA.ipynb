## Task 1

```python
import pandas as pd, numpy as np, matplotlib.pyplot  as plt, seaborn as sns
```

#### Load and read the data

```python
data = pd.read_csv('dataset.csv')
```

```python
data.head(5)
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isna().sum()
```

#### Removing rows with missing values

```python
data = data.dropna(subset=['Model'])
```

```python
data.isna().sum()
```

```python
data = data.dropna(subset=['Legislative District'])
```

```python
data.isna().sum()
```

```python
data = data.dropna(subset=['Vehicle Location','Electric Utility'])
```

```python
data.isna().sum()
```

#### Converting to category datatype

```python
data['Electric Vehicle Type'] = data['Electric Vehicle Type'].astype('category')
```

```python
data.info()
```

### Univariate analysis

#### Distribution of Car brand

```python

plt.figure(figsize=(15, 10))
plt.title('Distribution of Car brand')
barPlt = data['Make'].value_counts().plot(kind = 'bar', xlabel = 'Car brand', ylabel = 'count')

for i in barPlt.containers:
    barPlt.bar_label(i, label_type='edge')

plt.show()
```

#### Distribution of City

```python
plt.figure(figsize=(15, 10))
plt.title('Distribution of City')
barPlt = data['City'].value_counts().head(10).plot(kind = 'bar', xlabel = 'City', ylabel = 'frequency')

for i in barPlt.containers:
    barPlt.bar_label(i, label_type='edge')

plt.show()
```

#### Distribution of model year

```python
plt.figure(figsize=(8, 6))
plt.title('Distribution of model year')
barPlt = data['Model Year'].value_counts().plot(kind = 'line', xlabel = 'model year', ylabel = 'Number of Vehicles',marker = 'o')

plt.show()
```

#### Distribution of model

```python
plt.figure(figsize=(15, 10))
plt.title('Distribution of model')
barPlt = data['Model'].value_counts().head(10).plot(kind = 'bar', xlabel = 'Model', ylabel = 'frequency')

for i in barPlt.containers:
    barPlt.bar_label(i, label_type='edge')

plt.show()
```

#### Distribution of Electric Vehicle Types

```python
vehicle_type_counts = data['Electric Vehicle Type'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(vehicle_type_counts, labels=vehicle_type_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Electric Vehicle Types')
plt.show()
```

### Bivariate analysis

#### Electric Range by Electric Vehicle Type

```python
sns.violinplot(x='Electric Vehicle Type', y='Electric Range', data=data)
plt.title('Violin Plot of Electric Range by Electric Vehicle Type')
plt.show()
```

#### Top 10 cities with most Tesla cars

```python
data['Make'].mode()
```

```python
df = data[data['Make']=='TESLA']
```

```python
df = df['City'].value_counts().head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=df.index, y = df.values)
plt.title('Top 10 cities with most Tesla cars') 

plt.show()
```

## Task 2

```python
import plotly.express as px
```

```python
ev_cnt_pcode = data.groupby(['Postal Code', 'Model Year', 'State','City']).size().reset_index(name='EV_Count')
```

```python
fig = px.choropleth_mapbox(ev_cnt_pcode,
                           geojson="https://raw.githubusercontent.com/OpenDataDE/State-zip-code-GeoJSON/master/wa_washington_zip_codes_geo.min.json",
                           locations='Postal Code',
                           color='EV_Count',
                           featureidkey="properties.ZCTA5CE10",  
                           mapbox_style="carto-positron",
                           center={"lat": 47.6062, "lon": -122.3321}, 
                           title="Number of EV vehicles based on location Washington Over Year",
                           animation_frame="Model Year",  
                           color_continuous_scale="Cividis",
                           hover_data=['EV_Count']  
                          )


fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```
